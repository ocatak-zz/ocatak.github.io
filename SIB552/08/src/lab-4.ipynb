Bu lab çalışması kapsamında *PCA* ve *LDA* yöntemi ile boyut azaltma yapılacaktır.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
np.random.seed(26)
```

### Veri kümesi oluştur.

```python
X, y = make_classification(n_classes=3, n_samples=100000,n_features=100, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1, class_sep=1)
```

### PCA modeli oluştur.

```python
pca = PCA(n_components=3)
X_r = pca.fit(X).transform(X)
```

### LDA Modeli oluştur

```python
lda = LinearDiscriminantAnalysis(n_components=3)
X_r2 = lda.fit(X, y).transform(X)
```

### PCA - Sonucu görselleştir

```python
colors = ['navy', 'turquoise', 'darkorange']
for color, i in zip(colors, [0, 1, 2]):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, s=1)
plt.show()


```

### LDA - Sonucu görselleştir

```python
plt.figure()
for color, i in zip(colors, [0, 1, 2]):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color, s=1)
```
