Bu lab çalışmasında doğrusal ayrıştırılabilir bir veri kümesi kullanılarak SVM çekirdek fonksiyonları incelenecektir.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import MinMaxScaler
from matplotlib.colors import ListedColormap

cmap_light = ListedColormap(['#00AAAA', '#00FFAA', '#AAAAFF'])
```

Veri kümesi oluştur

```python
X, y = datasets.make_classification(n_samples=5000, n_features=2,
                                    n_informative=2, n_redundant=0,
                                    random_state=0)
```

Eğitim ve test veri kümesi oluştur

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

#### Doğrusal çekirdek fonksiyonu

```python
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=10).fit(X, y)
# create a mesh to plot in
x_min, x_max = X[:, 0].min() + 0.01, X[:, 0].max() + 0.01
y_min, y_max = X[:, 1].min() + 0.01, X[:, 1].max() + 0.01
h = (x_min / x_max)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max,0.01), np.arange(y_min, y_max,0.01))
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0],X[:, 1],c=y)
```

#### RBF çekirdek fonksiyonu

```python
plt.subplot(1, 1, 1)
C = 1.0 # SVM regularization parameter
svc2 = svm.SVC(kernel='rbf', C=1,gamma=10).fit(X, y)

Z = svc2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
#plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(X[:, 0],X[:, 1],c=y)
```

#### Polynomial çekirdek fonksiyonu

```python
plt.subplot(1, 1, 1)
C = 1.0 # SVM regularization parameter
svc2 = svm.SVC(kernel='poly', C=1,gamma=1, degree =5).fit(X, y)

Z = svc2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
#plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(X[:, 0],X[:, 1],c=y)
```
