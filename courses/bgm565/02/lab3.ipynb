Bu lab çalışmasında lojistik regresyon sınıflandırma yöntemini sentetik bir kümesine uygulayacağız.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# veri kumesini oku
verikumesi = pd.read_csv("ds_logreg.txt",delimiter="\t")
# ilk 10 satiri goster
verikumesi.head(10)
```

$X$ ve $y$ kullanılarak lojistik regresyon modelini eğit.

```python
X = verikumesi.iloc[:,:-1].values
y = verikumesi.iloc[:,X.shape[1]].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
# modeli tanimla
clf = LogisticRegression(verbose=0)
# Egitim veri kumesi kullanarak modeli egit
clf.fit(X_train, y_train)
```

```python
# agirliklar
clf.intercept_, clf.coef_
```

Sınıflandırma modelini kullanarak egitim veri kumesini tahmin et

```python
y_pred = clf.predict(X_test)
df = pd.DataFrame({"y":y_test,"y_pred":y_pred})
df
```
